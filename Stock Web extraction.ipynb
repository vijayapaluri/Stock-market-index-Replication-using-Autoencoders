## import packages

```python
#-------------Import packages-------------
import pandas as pd
import pandas_datareader as pdr #to extract data from yahoo finance
import numpy as np
import matplotlib.pyplot as plt
#from datetime import date
import time
```

## import data

```python
#-------------import data-------------
data = pd.read_csv(r"C:\Users\hp\Desktop\01ai_project\malasiya_stocks.csv")
```

```python
#store stock symbols into list
stocks_symbols = np.squeeze(data["Symbol"]).tolist()

# Index symbol for FTSE Bursa Malaysia KLCI on yahoo finance
index_symbol = '^KLSE;1=9'

#select time period
start_date = '2013-01-01'
#end_date =  date.today() # for todays date
end_date = '2021-12-05'

# Empty dataframe to store data from yahoo finance
data1 = pd.DataFrame() 
```

## extract index data from yahoo finance

```python
#-------------extract index data from yahoo finance-------------
# storing adjusted closing price of KLSE index
data1['KLSE'] = pdr.DataReader(index_symbol, 'yahoo', start_date, end_date)['Adj Close']

#-------------extract stock data from yahoo finance------------- 
#storing adjusted closing price of KLSE stocks
i = 0
while i < len(stocks_symbols):
    print('Downloading.... ', i, stocks_symbols[i])
    try:
        # Extract the desired data from Yahoo!
        data1[stocks_symbols[i]] = pdr.DataReader(stocks_symbols[i], 'yahoo', start_date, end_date)['Adj Close']
        i +=1      
    except:
        print ('Error with connection. Wait for 1 minute to try again...')
        time.sleep(30) # Wait for 30 seconds
        continue 
```

## data cleaning

```python
#-------------data cleaning-------------
data1.info()
data1.isna().sum()
# Remove columns with missing values and creating new dataframe
data2 = data1.drop(columns=['5285.KL','5296.KL'], axis =1)
data2.isna().sum()
```

## data visualization

```python
#-------------data visualization-------------
plt.figure('KLSE Index')
plt.legend()
plt.plot(data2.iloc[:,0])
plt.xlabel('Date(years)')
plt.ylabel('Adjusted price')
```
